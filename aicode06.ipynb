```
# 6.5.1. 계단함수

import numpy as np 
import matplotlib.pyplot as plt

def h(x):
    return np.array(x>0)

x = np.arange(-3, 3, 0.1)
y = h(x)
plt.plot(x,y)
plt.show()
```

```
# 6.5.5. 시그모이드 함수 1

import numpy as np 
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1+np.exp(-x))

x = np.arange(-5, 5, 0.1)
y = sigmoid(x)

plt.plot(x,y)
plt.show()
```

```
# 6.5.5. 시그모이드 함수 2

import numpy as np 
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1+np.exp(-x))

x = np.arange(-5, 5, 0.1)
y = sigmoid(x)

plt.figure(figsize=(8,6))
plt.xlim(-3,3)
plt.ylim(0,0.99)
plt.yticks(np.arange(0, 1.001, 0.1))
plt.xticks(np.arange(-5, 5.001, 1))

plt.plot(x,y)
plt.text(-2, .6, r'Groth  Curve', color='red')
plt.text(-2, .55, 'Sigmoid Curve', color='red')
plt.grid(dashes=(3,3),linewidth=0.5)

plt.show()
```

```
# 6.5.5. 시그모이드 함수 3

import numpy as np 
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1+np.exp(-x))


def h(x):
    return np.array(x>0)


x = np.arange(-5.0, -0.1, 0.1)
x = np.array(x.tolist()+[-0.000001,0.0,0.000001])
x = x.tolist()+np.arange(0.1, 5.0, 0.1).tolist()
x = np.array(x)

y = h(x)
plt.figure(figsize=(8,8))

plt.yticks(np.arange(0, 1.001, 0.1))
plt.xticks(np.arange(-6, 6.001, 2))
plt.grid(dashes=(3,3),linewidth=0.5)

plt.plot(x,y)

y = sigmoid(x)
plt.plot(x,y)

plt.show()
```

```
# 6.5.7	ReLU 1

import numpy as np 
import matplotlib.pyplot as plt

def r(x):
    return np.maximum(0, x)

x = np.arange(-5, 5, 0.1)
y = r(x)
plt.plot(x,y)
plt.show()
```
